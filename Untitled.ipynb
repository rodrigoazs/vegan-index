```python
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("brazil.csv")
```

```python
df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

X = df[["vegan", "vegetarian", "veg_options", "others"]]
y = df["total"]

regr = RandomForestRegressor(n_estimators=10000)
regr.fit(X, y)
```

```python
mean_absolute_error(y, regr.predict(X))
```

```python
regr.feature_importances_
```

```python
regr.feature_importances_
```

```python
m = X.to_numpy()
```

```python
(regr.feature_importances_ * m).sum(axis=1)
```

```python
df["vfi"] = (regr.feature_importances_ * m).sum(axis=1)
```

```python
s = df.sort_values(by=['vfi'])
s["city_index"] = s["city"].apply(lambda x: x.lower())
```

```python
s.head(20)
```

```python
ibge = pd.read_csv("municipios_ibge.csv", delimiter=";")
```

```python
ibge
```

```python
import unidecode
```

```python
ibge["Município"] = ibge["Município"].apply(lambda x: unidecode.unidecode(x))
```

```python
ibge = ibge[["Município", "População 2010"]]
ibge["População 2010"] = ibge["População 2010"].astype("float")
ibge.columns = ["city", "population"]
ibge["city_index"] = ibge["city"].apply(lambda x: x.lower())
```

```python
new_df = s.merge(ibge, how="left", on=["city_index"])
```

```python
new_df
```

```python
new_df[new_df.population.isna()].sort_values(by="city_index")["city_index"].tolist()
```

```python
new_df["vfi_pop"] = new_df["vfi"] / new_df["population"]
```
